```python
#this function checks all the mission values in the test and train data

def check_missing_values(train, test):
    print("="*40)
    print(f"(TRAIN) Number of missing values:")
    print("="*40)
    print(train.isna().sum()[train.isna().sum()>0])
    print("="*40)
    print(f"(TEST) Number of missing values:")
    print("="*40)
    print(test.isna().sum()[test.isna().sum()>0])    
    return

check_missing_values(df_test_b, df_train_b)  #calling the abve function
```

```python
# Function to perform EDA for each feature
def analyse_feature(feature_name):
    
    sns.jointplot(data=df_train_b, y="SalePrice", x=feature_name)  #setting up a joint plot between sale price and the feature
    plt.figure(figsize=(20, 30))
    plt.show()
    print("\n****INFO****")
    print(df_train_b[feature_name].describe()) #descriptive stats for the feature
    print("\n****VALUE COUNTS****")
    print(df_train_b[feature_name].value_counts(dropna=False)) #gives the value counts for the unique values in the feature
    print("\n****VALUE AVG SALE PRICE****")
    print(df_train_b.groupby(feature_name)['SalePrice'].mean().sort_values()) #mean sales price for unique value of the feature
    if df_train_b[feature_name].dtype!="O":
        print("\nSkewness:",str(skew(df_train_b[feature_name]))) #checkes the skewness of the feature

    print("\n****TEST INFO****")
    print(df_test_b[feature_name].describe()) #descriptive stats for the feature
    print("\n****VALUE COUNTS****")
    print(df_test_b[feature_name].value_counts()) #gives the value counts for the unique values in the feature
    
    #checks if the data is only in train DF
    print("\nOnly in Train: "+str(list(set(df_train_b[feature_name].value_counts().index.values) - set(df_test_b[feature_name].value_counts().index.values))))
    #checks if the data is only in train DF
    print("Only in Test: "+ str(list(set(df_test_b[feature_name].value_counts().index.values) - set(df_train_b[feature_name].value_counts().index.values))))
    
analyse_feature("Neighborhood")  #enter the feature name to look at the EDA
```

- New features are created.

```python
#Feature Extraction (Converting Years to Ages)
df_train_b["Age"] = 2011 - df_train_b["YearBuilt"]
df_test_b["Age"] = 2011 - df_test_b["YearBuilt"]

df_train_b["RemodAfter"] = df_train_b["YearRemodAdd"] - df_train_b["YearBuilt"]
df_train_b["AgeRemodAdd"] = 2011 - df_train_b["YearRemodAdd"]

df_test_b["RemodAfter"] = df_test_b["YearRemodAdd"] - df_test_b["YearBuilt"]
df_test_b.loc[df_test_b.RemodAfter<0,'RemodAfter'] = 0
df_test_b["AgeRemodAdd"] = 2011 - df_test_b["YearRemodAdd"]

df_train_b["Age_Sold"] = df_train_b["YrSold"] - df_train_b["YearBuilt"] 
df_test_b["Age_Sold"] = df_test_b["YrSold"] - df_test_b["YearBuilt"]
df_test_b.loc[df_test_b.Age_Sold<0,'Age_Sold'] = 0

df_train_b["Sold_before"] = 2011 - df_train_b["YrSold"]
df_test_b["Sold_before"] = 2011 - df_test_b["YrSold"] 

df_train_b['Age_Garage'] = 2011 - df_train_b['GarageYrBlt']
df_test_b['Age_Garage'] = 2011 - df_test_b['GarageYrBlt']
df_train_b['Age_Garage_Sold'] = df_train_b["YrSold"] - df_train_b['GarageYrBlt']
df_test_b['Age_Garage_Sold'] = df_test_b["YrSold"] - df_test_b['GarageYrBlt']

#adding all the porches together to get totalporch
df_train_b['TotalPorch']=df_train_b['EnclosedPorch'] + df_train_b['OpenPorchSF'] + df_train_b['ScreenPorch'] + df_train_b['3SsnPorch'] 
df_test_b['TotalPorch']=df_test_b['EnclosedPorch'] + df_test_b['OpenPorchSF'] + df_test_b['ScreenPorch'] + df_test_b['3SsnPorch'] 

#adding all the baths together 
df_train_b['TotalBath'] = df_train_b['FullBath']+ (df_train_b['HalfBath']*0.5)+ df_train_b['BsmtFullBath']+ (df_train_b['BsmtHalfBath']*0.5)
df_test_b['TotalBath'] = df_test_b['FullBath']+ (df_test_b['HalfBath']*0.5)+ df_test_b['BsmtFullBath']+ (df_test_b['BsmtHalfBath']*0.5)
```

```python
df_train_b["Has2ndFlr"] = df_train_b["2ndFlrSF"].apply(lambda x: "Yes" if x>0 else "No") #check if there is 2nd floor in the house
df_test_b["Has2ndFlr"] = df_test_b["2ndFlrSF"].apply(lambda x: "Yes" if x>0 else "No")   #check if there is 2nd floor in the house

df_train_b["HasBsmt"] = df_train_b['TotalBsmtSF'].apply(lambda x: "Yes" if x>0 else "No") #check if there is bsmt in the house
df_test_b["HasBsmt"] = df_test_b['TotalBsmtSF'].apply(lambda x: "Yes" if x>0 else "No")   #check if there is bsmt in the house
```

```python
def random_forest_model(X, y):
    """
    Trains a random forest regressor on the given data and returns the model, RMSE score and model type.

    Args:
        X (pandas.DataFrame): The feature matrix.
        y (pandas.Series): The target variable.

    Returns:
        Tuple: A tuple containing the model, RMSE score, and model type.
    """
    # split data
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=219)
    
    # create and fit model
    model = RandomForestRegressor(n_estimators=100, random_state=420)
    model.fit(x_train, y_train)
    
    # get model predictions
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)

    # evaluate predictions
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
    print(f"Training RMSE: {train_rmse:.3f}")
    print(f"Test RMSE: {test_rmse:.3f}")
  

    
    return 'RandomForest', test_rmse, 'non-parametric'
        
ret_model, ret_score, ret_model_type  = random_forest_model(df_nridght_skewed, y_nridght_log)

model.append(ret_model)
score.append(ret_score)
model_type.append(ret_model_type)
```

```python
def neighborhood_reg_model(X, y):
    """
    Trains a RandomForestRegressor model on the input data and returns the top 15 important features.

    Args:
    X: input feature data
    y: output label data

    Returns:
    A pandas DataFrame containing the top 15 important features and their corresponding feature coefficients.
    """
    
    # split data
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=219)
    
    # create and fit model
    regression_model = RandomForestRegressor()
    regression_model.fit(x_train, y_train)
    
    # feat feature names and importance
    feature_names = df_nridght_skewed.columns
    feature_importance_abs = np.abs(regression_model.feature_importances_)
    feature_importance = regression_model.feature_importances_
    
    # create feature importance dictionary
    sorted_idx = np.argsort(feature_importance)[::-1]
    sorted_feature_importance_dict = {}
    for i in sorted_idx:
        sorted_feature_importance_dict[feature_names[i]] = [feature_importance[i]]
        
    # create absolute feature importance dictionary
    sorted_idx = np.argsort(feature_importance_abs)[::-1]
    sorted_feature_importance_dict_abs = {}
    for i in sorted_idx:
        sorted_feature_importance_dict_abs[feature_names[i]] = [feature_importance_abs[i]]
    
    # rearrange dictionary 
    sorted_feature_importance_dict_rearanged = {}
    for i in sorted_feature_importance_dict_abs:
        sorted_feature_importance_dict_rearanged[i] = sorted_feature_importance_dict[i]
    
    return pd.DataFrame.from_dict(sorted_feature_importance_dict_rearanged, orient='index', columns=['FeatureCoef'])[:15]

        
df_nridght_imp_features = neighborhood_reg_model(df_nridght_skewed, y_nridght_log)
df_nridght_imp_features
```

### Model for Complete Train DataFrame (kaggle Model w/o Hyper Parameters)

```python
class ModelTuner:
    def __init__(self, models, X, y, cv, loss):
        """
        Initializes a model trainer object with the given parameters.

        :param models: a list of machine learning models to train
        :param X: the feature matrix for the training data
        :param y: the target variable for the training data
        :param cv: the number of cross-validation folds to use for training and evaluation
        :param loss: the loss function to use for model evaluation
        """
        # Make a copy of the initial models list and store it
        self.init_models = models.copy()
        self.models = models.copy()

        # Store the training data and target variable
        self.X_train = df_train_b_skewed.copy()
        self.y_train = y_log.copy()

        # Store the number of cross-validation folds and loss function
        self.cv = cv
        self.loss = loss

        # Initialize dictionaries to store training and validation scores for each model
        self.train_scores = {}
        self.valid_scores = {}
        
    def tune_model(self, model_name, grid, verbose=1): #n_jobs=-1):
        """
        Trains and tunes a specified machine learning model using a grid search over hyperparameters.

        :param model_name: the name of the model to tune
        :param grid: a dictionary of hyperparameters to search over using a grid search
        :param verbose: the level of verbosity for the grid search (default 1)
        :return: the trained and tuned machine learning model
        """
        # Set the model's hyperparameters to the specified grid and train the model
        self.models[model_name] = self.init_models[model_name]
#         search = GridSearchCV(
#             self.models[model_name],
#             param_grid=grid,
#             scoring=self.loss,
#             cv=self.cv,
#             return_train_score=True,
#             verbose=verbose,
#             n_jobs=n_jobs,
#         )
#         search.fit(self.X_train, self.y_train)
#         self.models[model_name] = search.best_estimator_
       
        model = self.models[model_name].set_params(**grid)
        model.fit(self.X_train, self.y_train)
        rename_metric = {
            "mean_train_score": "mean_train_RMSE",
            "mean_test_score": "mean_valid_RMSE",
        }        
        
#         cv_results = pd.DataFrame(search.cv_results_)
#         columns = [f"param_{param}" for param, val in grid.items() if len(val)>1]
#         columns.extend(rename_metric.keys())
#         cv_results = cv_results[columns].rename(columns=rename_metric)
#         cv_results[list(rename_metric.values())] = -cv_results[list(rename_metric.values())]
#         cv_results = cv_results.sort_values(by="mean_valid_RMSE", ignore_index=True)
    
#         train_score = cv_results.mean_train_RMSE.iloc[0]
#         valid_score = cv_results.mean_valid_RMSE.iloc[0]
#         self.train_scores[model_name] = train_score
#         self.valid_scores[model_name] = valid_score
        
#         print("="*40)
#         print(f"Model: {search.best_estimator_}")
#         print(f"Train RMSE: {train_score}")
#         print(f"Valid RMSE: {valid_score}")
#         print("="*40)

#         return cv_results
        
        # Compute the training and validation RMSE scores for the model  
        train_rmse = np.sqrt(mean_squared_error(self.y_train, model.predict(self.X_train)))
        valid_scores = cross_validate(
            model,
            X=self.X_train,
            y=self.y_train,
            cv=self.cv,
            return_train_score=False,
            scoring=self.loss,
        )["test_score"]
        valid_rmse = -np.mean(valid_scores)
        
        # Update the trained model, training scores, and validation scores in the object's attributes
        self.models[model_name] = model
        self.train_scores[model_name] = train_rmse
        self.valid_scores[model_name] = valid_rmse
        
        # Print the model and its training and validation scores
        print("="*40)
        print(f"Model: {model}")
        print(f"Train RMSE: {train_rmse}")
        print(f"Valid RMSE: {valid_rmse}")
        print("="*40)
        
        # Return the trained and tuned model
        return model
    
    def collate_results(self):
        """
        Combines the training and validation RMSE scores for each trained machine learning model into a single Pandas DataFrame.

        :return: a Pandas DataFrame containing the training and validation RMSE scores for each trained machine learning model
        """
        # Create Pandas DataFrames for the training and validation RMSE scores
        train_results = pd.DataFrame.from_dict(
            self.train_scores,
            orient="index",
            columns=["mean_train_RMSE"],
        )
        valid_results = pd.DataFrame.from_dict(
            self.valid_scores,
            orient="index",
            columns=["mean_valid_RMSE"],
        )
        
        # Concatenate the training and validation RMSE DataFrames horizontally and sort by validation RMSE
        
        results = pd.concat(
            [train_results, valid_results],
            axis=1,
        ).sort_values(by="mean_valid_RMSE")
        
        # Return the concatenated results DataFrame
        return results
    
    def get_models(self, model_names=None):
        """
        Returns a list of trained machine learning models.

        :param model_names: a list of model names to return; if None (default), 
                            all trained models are returned
        :return: a list of tuples, where each tuple contains a model name and
                 its corresponding trained machine learning model object
        """
        # If no model names are provided, use all trained models
        if not model_names:
            model_names = self.models.keys()
        # Create a list of tuples, where each tuple contains a model name and its corresponding trained machine learning model object    
        models = [(name, self.models[name]) for name in model_names]
        
        # Return the list of model tuples
        return models
    
    def run_cv(self, model, n_jobs=-1):
        """
        This method runs cross-validation for a specified model.

        Args:
        model (str or estimator object): The name of the model in the self.models dict or the estimator object.
        n_jobs (int, optional): Number of CPU cores used for parallelization. Default is -1, which means to use all available CPUs.

        Returns:
        None. The method prints the mean train RMSE and mean valid RMSE of the model.
        """
        # If model is a string, get the corresponding estimator object from the models dictionary
        model = self.models[model] if model in self.models else model
        
         # Run cross-validation on the model using the specified number of CPUs
        scores = cross_validate(
            model,
            X=self.X_train,
            y=self.y_train,
            cv=self.cv,
            return_train_score=True,
            scoring=self.loss,
            n_jobs=n_jobs,
        )
        # Extract the train and validation scores
        train_scores = scores["train_score"]
        valid_scores = scores["test_score"]
        
        # Print the results
        print(f"Model: {model}")
        print(f"Train RMSE: {-np.mean(train_scores)}")
        print(f"Valid RMSE: {-np.mean(valid_scores)}")

    
    def model_predict(self, model, X_test):
        """
        Predicts the sale price for the test dataset using the specified model.

        Args:
            model (str or sklearn model): The name of the model or a sklearn model instance to use for prediction.
            X_test (pd.DataFrame): The test dataset to use for prediction.

        Returns:
            pd.DataFrame: A dataframe containing the predicted sale prices.
        """
        # Get the model instance
        model = self.models[model] if model in self.models else model
        
        # Fit the model on the training dataset
        model.fit(self.X_train, self.y_train)
        
        # Predict the sale prices for the test dataset
        y_pred = model.predict(X_test)
        
        # Convert the sale prices to the correct format and create a dataframe
        y_pred = pd.DataFrame({
            "Id": range(1461, 1461+len(y_pred)),
            "SalePrice": np.expm1(y_pred),
        })

        return y_pred
    
    def perf_boxplot(self, add_models=None, cv=None, n_jobs=-1):
        """
        Creates a box plot showing the distribution of validation RMSE scores for each model.

        Parameters:
        - add_models: a dictionary of additional models to include in the plot, in the format {model_name: model_instance}
        - cv: the cross-validation strategy to use. If None, uses the strategy set in the constructor.
        - n_jobs: the number of parallel jobs to run for cross-validation. -1 means using all processors.

        Returns: None
        """
    
        # Combine the main models and the additional models, if any
        models = self.models.copy()
        if add_models:
            for name, model in add_models:
                models[name] = model

        # Collect the validation scores for each model
        perf_df = pd.DataFrame()
        for name, model in models.items():
            valid_scores = cross_validate(
                model,
                X=self.X_train,
                y=self.y_train,
                cv=cv if cv else self.cv,
                scoring=self.loss,
                n_jobs=n_jobs,
            )["test_score"]

            # Store the validation scores for the model in a dataframe
            model_perf = pd.DataFrame({"Model": name, "Validation RMSE": -valid_scores})
            perf_df = pd.concat([perf_df, model_perf])

        # Calculate the minimum median RMSE among all models
        min_median = perf_df.groupby("Model").median().min()[0]

        # Plot the box plot
        plt.figure(figsize=(18, 8))

        sns.boxplot(
            x=perf_df["Validation RMSE"],
            y=perf_df["Model"],
            flierprops={"alpha": 0.5},
        )
        plt.axvline(x=min_median, ls="--", color="coral")
        plt.ylabel("")
```

```python
# define the number of folds for KFold cross-validation
kfold = KFold(n_splits=10, shuffle=True, random_state=SEED)

# define the loss function for scoring
rmse_loss = make_scorer(mean_squared_error, squared=False, greater_is_better=False)

# define the models to be tuned
MODELS = {
    "Ridge": Ridge(random_state=SEED),
    "Linear SVR": LinearSVR(random_state=SEED),
    "Kernel Ridge": KernelRidge(),
    "Kernel SVR": SVR(),
    "Random Forest": RandomForestRegressor(random_state=SEED),
    "Gradient Boosting": GradientBoostingRegressor(random_state=SEED),
}

# create an instance of the ModelTuner class
tuner = ModelTuner(MODELS, X=df_train_b_skewed, y=y_log, cv=kfold, loss=rmse_loss)
```
